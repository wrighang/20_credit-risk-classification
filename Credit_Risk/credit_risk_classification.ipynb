```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
file_path = Path("./Resources/lending_data.csv")
lending_data_df = pd.read_csv(file_path)

'''A value of 0 in the “loan_status” column means that the loan is healthy. 
A value of 1 means that the loan has a high risk of defaulting.'''

# Review the DataFrame
lending_data_df
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels/target
y = lending_data_df['loan_status']

# Separate the X variable, the features
X = lending_data_df.drop(columns=['loan_status'])
```

```python
#count how many rows 
len(X)
```

```python
# Review the y variable Series
display(y.head())
display(y.tail())
```

```python
# Review the X variable DataFrame
X
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1,
                                                    stratify=y)
X_train.shape
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
logistic_regression_model = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)


# Fit the model using training data
fitted_model = logistic_regression_model.fit(X_train, y_train)
logistic_regression_model
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
testing_predictions = logistic_regression_model.predict(X_test)

#look at training predictions
training_predictions = fitted_model.predict(X_train)


print(f"Training Data Score: {logistic_regression_model.score(X_train, y_train)}")
print(f"Testing Data Score: {logistic_regression_model.score(X_test, y_test)}")


results_df = pd.DataFrame({"Prediction": testing_predictions, "Actual": y_test}).tail(10)
results_df
```

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
from sklearn.metrics import accuracy_score
acc_score_test = accuracy_score(y_test, testing_predictions)
acc_score_test
```

```python
acc_score_train = accuracy_score(y_train, training_predictions)
acc_score_train
```

```python
#generate a confusion matrix for the model (training matrix)
cm_train = confusion_matrix(y_train, training_predictions)
print(cm_train)

cm_train_df = pd.DataFrame(
    cm_train, index=["actual_0_healthy_loan", "actual_1_high_risk_loan"], columns=["predicted_0_healthy_loan", "predicted_1_high_risk_loan"]
)
cm_train_df
```

```python
# Generate a confusion matrix for the model (testing data)
cm_test = confusion_matrix(y_test, testing_predictions)
print(cm_test)

cm_test_df = pd.DataFrame(
    cm_test, index=["actual_0_healthy_loan", "actual_1_high_risk_loan"], columns=["predicted_0_healthy_loan", "predicted_1_high_risk_loan"]
)
cm_test_df
```

```python
# Print the classification report for the model (training classification report)
target_names = ["healthy_loan", "high_risk_loan"]
class_report_train = (classification_report(y_train, training_predictions, target_names=target_names))
```

```python
# Print the classification report for the model (testing classification report)
target_names = ["healthy_loan", "high_risk_loan"]
class_report_test = (classification_report(y_test, testing_predictions, target_names=target_names))
```

```python
#display results for train 
print("Confusion Matrix - Train")
display(cm_train_df)
print(f"Accuracy Score : {acc_score_train}")
print("Classification Report")
print(class_report_train)
```

```python
#display results for test
print("Confusion Matrix - Test")
display(cm_test_df)
print(f"Accuracy Score : {acc_score_test}")
print("Classification Report")
print(class_report_test)
```
